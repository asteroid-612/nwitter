    - 카테고리를 제외하고 구성하는 column의 숫자를 맞춰서 적용해줘야함.(option 에서 적용해줄때 그 카테고리의 숫자에 맞춰야함.)

```python
%%html
<html>
  <head>
    <script type="text/javascript" src="https://www.gstatic.com/charts/loader.js"></script>
    <script type="text/javascript">
      google.charts.load('current', {'packages':['corechart']});
      google.charts.setOnLoadCallback(initialize);
    
      function initialize(){
          $('#btn').click(function(){
              drawVisualization();
          })
      }
        
     function themeSelection(theme){
         if (theme == "Sky"){
             return colorChart[0];
         }
         else if(theme == "Sand"){
             return colorChart[1];
         }
         else if(theme == "2018"){
             return colorChart[3];
         }
         else if(theme == "2017"){
             return colorChart[4];
         }
         else{
             return colorChart[2];
         }
     }
        
      var colorChart = [
          ['#bccad6','#8d9db6','#667292','#f1e3dd','#bccad6'],
          ['#fbefcc', '#f9ccac', '#f4a688', '#e0876a', '#fbefcc'],
          ['#96ceb4', '#ffeead', '#ffcc5c','#ff6f69','#96ceb4'],
          ['#6b5b95', '#feb236', '#d64161','#ff7b25','#6b5b95'],
          ['#d6cbd3', '#eca1a6', '#bdcebe', '#ada397', '#d6cbd3'],
          
      ];

      function drawVisualization() {
        // Some raw data (not necessarily accurate)
        var data = google.visualization.arrayToDataTable([
          ['Category', '안전한 장소', '위험한 장소', '공공시설', '공사중인 장소', '부끄러운 장소', '평균치'],
          ['2019/05',  165,      938,         522,             998,           450,      614.6],
          ['2019/06',  135,      1120,        599,             1268,          288,      682],
          ['2019/07',  157,      1167,        587,             807,           397,      623],
          ['2019/08',  139,      1110,        615,             968,           215,      609.4],
          ['2019/09',  136,      691,         629,             1026,          366,      569.6]
        ]);

        var in_title = $('#chart_title').val();
        var text_color = $('#color_select>option:selected').text();
        var text_font = $('#font_select>option:selected').text();
        var text_size = $('#font_size').val();
        var theme = themeSelection(String($('#color_palette>option:selected').text()));
        var options = {
          title : in_title,
          titleTextStyle : {
              color : text_color,
              fontName : text_font,
              fontSize : text_size,
          },
          vAxis: {title: 'Result'},
          hAxis: {title: 'Month'},
          seriesType: 'bars',
          series: {5: {type: 'line'}},
          colors: theme,
        };  

        var chart = new google.visualization.ComboChart(document.getElementById('chart_div'));
        chart.draw(data, options);
      }
    </script>
  </head>
  <body>
    <div>
        Title : <input type="text" name="chart_title" value="" id="chart_title">
    </div>
    <div>
        Title Font select : <select name="job" id="font_select">
                        <option value="">Times New Roman</option>
                        <option value="">Helvetica</option>
                        <option value="">Arial Black</option>
                        <option value="">Comic Sans MS</option>
                      </select>
    </div>
    <div>
        Title Font size select : <input type="text" name="font_size" value="" id="font_size">
    </div>
    <div>
        Title Color select : <select name="color" id="color_select">
                            <option value="">Red</option>
                            <option value="">Blue</option>
                            <option value="">Yellow</option>
                            <option value="">Black</option>
                        </select>
    </div>
    <div>
        Chart Color Theme Select : <select name="palette" id="color_palette">
                                <option value="">Sky</option>
                                <option value="">Sand</option>
                                <option value="">Beach</option>
                                <option value="">2018</option>
                                <option value="">2017</option>
                        </select>
    </div>
    <div id="chart_div" style="width: 900px; height: 500px;">
        <button id="btn" type="button" onclick="">draw</button>
    </div>
  </body>
</html>
```
