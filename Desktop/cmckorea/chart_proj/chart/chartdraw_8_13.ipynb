```python
with open('kookmindust.json') as json_file:
    json_data = json.load(json_file)
    
# print(json_data['dataList'][0]['var2'])
# print(json_data['dataList'][0]['category'])
# print(json_data['dataList'][0]['lat'])
# print(json_data['dataList'][0]['lon'])
# print(json_data['dataList'][0]['name'])

#for i in range(0, len(json_data['dataList'])):
#    print(json_data['dataList'][i]['oid'])
```

```python
c_data = {}

for i in range(0, len(json_data['dataList'])):
    p_dict = {}
    p_dict['writer_name'] = json_data['dataList'][i]['var2']
    p_dict['category'] = json_data['dataList'][i]['category']
    p_dict['lat'] = json_data['dataList'][i]['lat']
    p_dict['lon'] = json_data['dataList'][i]['lon']
    p_dict['loc_name'] = json_data['dataList'][i]['name']
    oid = json_data['dataList'][i]['oid']
    c_data[oid] = p_dict

print(c_data)
```

```python
with open('chart_data.json','w') as outfile:
    json.dump(c_data, outfile)
```
